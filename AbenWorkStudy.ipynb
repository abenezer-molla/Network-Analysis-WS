```python
print(__doc__)

import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas

import argparse

import re

ap = argparse.ArgumentParser()
ap.add_argument("-p", "--file_path", default = 'WorkStudy/',
                help="path where all your csv files may be. \
                Please update the default with your common folder.")
ap.add_argument("-f", "--filename", default = 'MMG_sector_groups.csv')

args = vars(ap.parse_args())

x = pandas.read_csv('MMG_sector_groups.csv', usecols = ['macro_sector'])


booleans = []
for result in x.macro_sector:
    if re.search('Education', result):
        booleans.append(True)
    elif re.search('Health', result):
        booleans.append(True)       
    elif re.search('Wash', result):        
        booleans.append(True)        
    else:
        booleans.append(False)

print(len(booleans))

Filtered = pd.Series(booleans)

NewData = x[Filtered]

NewData.shape

Rows_Removed = x.shape[0] - NewData.shape[0]

print("Rows Removed = ", str(Rows_Removed))
print(x)

columns = int(input('How many columns?:'))

for i in range(columns):
    plt.figure(NewData.columns[i]); plt.hist(NewData.iloc[:,i]); plt.show()
```
